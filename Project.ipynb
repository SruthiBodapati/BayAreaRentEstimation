```python
##importing csv files after scrapping  and changing into desired data frames

import csv
import pandas as pd
pd.options.display.max_rows = 10000
df=pd.read_csv(r'C:\Users\sruth\Desktop\second sem\Data Analytics\project\SFBayArea_Rental.csv')

Oakland_Data=df[df['Address'].str.contains('Oakland, CA', case=False)]
Fremont_Data=df[df['Address'].str.contains('Fremont, CA', case=False)]
Fremont_Data.index -= 565
Oakland_Data.index += 1
del Fremont_Data['Unnamed: 0']
del Oakland_Data['Unnamed: 0']
Oakland_Data #Oakland city Data frame
Fremont_Data #fremont city Data frame

```

```python
##checking for null values and sizes

print(Fremont_Data.info())
print(Fremont_Data.shape)
print(Oakland_Data.info())
print(Oakland_Data.shape)

```

```python
##Cleaning and pre processing the data for Fremont

#Rent
pd.options.mode.chained_assignment = None
Fremont_Data = Fremont_Data[Fremont_Data['Rent'].str.contains('', case=False)]
Fremont_Data["Rent"]= Fremont_Data["Rent"].str.replace("/mo", "") 
Fremont_Data["Rent"]= Fremont_Data["Rent"].str.replace("$", "") 
Fremont_Data["Rent"]= Fremont_Data["Rent"].str.replace(",", "") 
pd.to_numeric(Fremont_Data["Rent"])
Fremont_Data["Rent"]= Fremont_Data["Rent"].apply(pd.to_numeric)

#Size
Fremont_Data = Fremont_Data[Fremont_Data['Size'].str.contains('', case=False)]
Fremont_Data["Size"]= Fremont_Data["Size"].str.replace("sqft", "") 
#Fremont_Data["Size"]= Fremont_Data["Size"].str.replace("$", "") 
Fremont_Data["Size"]= Fremont_Data["Size"].str.replace(",", "") 
pd.to_numeric(Fremont_Data["Size"])
Fremont_Data["Size"]= Fremont_Data["Size"].apply(pd.to_numeric)

#Income
Fremont_Data = Fremont_Data[Fremont_Data['Income'].str.contains('', case=False)]
Fremont_Data["Income"]= Fremont_Data["Income"].str.replace("$", "") 
Fremont_Data["Income"]= Fremont_Data["Income"].str.replace(",", "") 
pd.to_numeric(Fremont_Data["Income"])
Fremont_Data["Income"]= Fremont_Data["Income"].apply(pd.to_numeric)

#beds
Fremont_Data = Fremont_Data[Fremont_Data['Bed'].str.contains('', case=False)]
Fremont_Data["Bed"]= Fremont_Data["Bed"].str.replace("bd", "") 
pd.to_numeric(Fremont_Data["Bed"])
Fremont_Data["Bed"]= Fremont_Data["Bed"].apply(pd.to_numeric)

#baths
Fremont_Data = Fremont_Data[Fremont_Data['Bath'].str.contains('', case=False)]
Fremont_Data["Bath"]= Fremont_Data["Bath"].str.replace("ba", "") 
pd.to_numeric(Fremont_Data["Size"])
Fremont_Data["Size"]= Fremont_Data["Size"].apply(pd.to_numeric)
```

```python
##Cleaning and preprocessing the data for Oakland_Data

#Rent
pd.options.mode.chained_assignment = None
Oakland_Data = Oakland_Data[Oakland_Data['Rent'].str.contains('', case=False)]
Oakland_Data["Rent"]= Oakland_Data["Rent"].str.replace("/mo", "") 
Oakland_Data["Rent"]= Oakland_Data["Rent"].str.replace("$", "") 
Oakland_Data["Rent"]= Oakland_Data["Rent"].str.replace(",", "") 
pd.to_numeric(Oakland_Data["Rent"])
Oakland_Data["Rent"]= Oakland_Data["Rent"].apply(pd.to_numeric)

#Size
Oakland_Data = Oakland_Data[Oakland_Data['Size'].str.contains('', case=False)]
Oakland_Data["Size"]= Oakland_Data["Size"].str.replace("sqft", "") 
#Oakland_Data["Size"]= Oakland_Data["Size"].str.replace("$", "") 
Oakland_Data["Size"]= Oakland_Data["Size"].str.replace(",", "") 
pd.to_numeric(Oakland_Data["Size"])
Oakland_Data["Size"]= Oakland_Data["Size"].apply(pd.to_numeric)

#Income
Oakland_Data = Oakland_Data[Oakland_Data['Income'].str.contains('', case=False)]
Oakland_Data["Income"]= Oakland_Data["Income"].str.replace("$", "") 
Oakland_Data["Income"]= Oakland_Data["Income"].str.replace(",", "") 
pd.to_numeric(Oakland_Data["Income"])
Oakland_Data["Income"]= Oakland_Data["Income"].apply(pd.to_numeric)

#beds
Oakland_Data = Oakland_Data[Oakland_Data['Bed'].str.contains('', case=False)]
Oakland_Data["Bed"]= Oakland_Data["Bed"].str.replace("bd", "") 
pd.to_numeric(Oakland_Data["Bed"])
Oakland_Data["Bed"]= Oakland_Data["Bed"].apply(pd.to_numeric)

#baths
Oakland_Data = Oakland_Data[Oakland_Data['Bath'].str.contains('', case=False)]
Oakland_Data["Bath"]= Oakland_Data["Bath"].str.replace("ba", "") 
pd.to_numeric(Oakland_Data["Size"])
Oakland_Data["Size"]= Oakland_Data["Size"].apply(pd.to_numeric)
```

```python
##Summary of data columns of Fremont and Oakland cities regarding rent, beds, size and Income respectively.

print(Fremont_Data.describe())
print(Oakland_Data.describe())

```

```python
#Checking the outliers using box plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.rcParams["axes.grid"] = False
fig=plt.figure(figsize=(10,10))
f1=fig.add_subplot(2,2,1)  
f2=fig.add_subplot(2,2,2)
f1.boxplot(x=Fremont_Data["Rent"],vert=False)
f2.boxplot(x=Oakland_Data["Rent"],vert=False)
f1.set_xlabel('Fremont outliers')
f2.set_xlabel('Oakland outliers')
fig.suptitle('Outliers')

plt.show()

```

```python
bins = numpy.linspace(0, 10000, 25)
plt.hist(Fremont_Data["Rent"], bins, alpha=0.9, label='Fremont')
plt.hist(Oakland_Data["Rent"], bins, alpha=0.8, label='Oakland')
plt.legend(loc='upper right')
plt.title('plotting rents')
plt.xlabel('Rent rate')
plt.ylabel('Properties range')
plt.show()
```

```python
# libraries

 
o=pd.pivot_table(Oakland_Data,index=['Bed'],columns=['Address'], values="Rent",aggfunc=np.mean)
f=pd.pivot_table(Fremont_Data,index=['Bed'],columns=['Address'], values="Rent",aggfunc=np.mean)
fig, (ax1, ax2) = plt.subplots(2)

ax2.set_ylim([0,8000])
ax1.set_ylim([0,8000])
o.plot(ax=ax1)
f.plot(ax=ax2)
ax1.set_title('Rent vs Beds')
ax1.set_xlabel('Bed')
ax1.set_ylabel('Rent')
ax2.set_xlabel('Bed')
ax2.set_ylabel('Rent')
ax1.get_legend().remove()
ax2.get_legend().remove()



#Inferences
# Fremont- The rents are not scattered across different communities
# Oakland - There is a lot of variance in rents across different communities. This can mean there is a wealth gap. 
# There may be some poor neiborhoods
```

```python
#Commute88% of residents commute by car.

Fremont_Data['Commute_rate'] = Fremont_Data['Commute'].str.extract('(\d+)', expand=False).str.strip()
Oakland_Data['Commute_rate'] = Oakland_Data['Commute'].str.extract('(\d+)', expand=False).str.strip()

Fremont_Data['city'] = "Fremont"
Oakland_Data['city'] = "Oakland"
res_merge = pd.concat([Fremont_Data, Oakland_Data])
#pd.to_numeric(Oakland_Data["Size"])
res_merge["Commute_rate"]= res_merge["Commute_rate"].apply(pd.to_numeric)
res_merge
res_merge.to_csv('Fremont_Oakland_Data.csv')
```

```python
plt.scatter(res_merge['city'], res_merge['Commute_rate'], marker='8')
plt.xlabel("Cities")
plt.ylabel("Commute rate")
plt.title("Commute comparsion between two cities")

#Inferences
#Majority of Fremont residents commute by car
#Oakland residents relatively travel less by car, and also uses other means of transporation
```
